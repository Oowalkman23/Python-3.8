```python
# Necessary initial library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Grab data, example = Korea GDP dataset
dataset = pd.read_excel('D:\Belajar\Project Py\Dataset\korea_data.xlsx')
dataset.head()
dataset.columns
```

```python
# Rename the columns
dataset.rename(
    columns={'GDP growth (annual %)':'GDP growth',
                        'Gross capital formation (% of GDP)':'Gross cap formation',
                        'Population growth (annual %)':'Pop growth',
                        'Birth rate, crude (per 1,000 people)':'Birth rate',
                        'Gross capital formation (annual % growth)':'Gc formation annual',
                        'Broad money growth (annual %)':'Broad money growth',
                        'Final consumption expenditure (annual % growth)':'Final consumption expenditure',
                        'General government final consumption expenditure (annual % growth)':'Government final consumption expenditure',
                        'Households and NPISHs Final consumption expenditure (annual % growth)':'Hh final consumption expenditure',
                         'Unemployment, total (% of total labor force) (national estimate)':'Unemployment'},     inplace=True)

dataset.head()
```

```python
# Check for nulls
dataset.isnull().any()
# Remove year 2017 due to '..' values

#Create new dataset, drop 2017
dataset_new = dataset.drop(48, axis=0)
dataset_new.set_index('Year', drop=True, inplace=True)
dataset_new = dataset_new.astype(float)
dataset_new
```

```python
# Check for multicollinearity (Multivariate Linear Regression Assumption)
correlation = dataset_new.corr()
correlation

# For relation of 2 different variable, closer to 1 or -1 means more collinear between those variables which is bad for regression
```

```python
# For different perspective, seaborn.heatmap will do
import seaborn as sbn
sbn.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, cmap='RdBu')
```

```python
# Another method is to use variance inflation factor provided in statsmodels library
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

x_now = sm.tools.add_constant(dataset_new)

series_new = pd.Series([variance_inflation_factor(x_now.values, i) for i in range(x_now.shape[1])], index = x_now.columns)

display(series_new)
```

```python
# Create new dataset after dropping column with variance inflation factor > 10, except the dependent variable, GDP Growth.
dataset_new_after = dataset_new.drop(['Birth rate', 'Final consumption expenditure', 'Government final consumption expenditure'], axis=1)

# Check the variance inflation factor again
x_new = sm.tools.add_constant(dataset_new_after)
series_after = pd.Series([variance_inflation_factor(x_new.values, i) for i in range(x_new.shape[1])], index = x_new.columns)

series_after
```

```python
# Visualize by scatter plot for distribution in data_set_new
pd.plotting.scatter_matrix(dataset_new_after, alpha = 1, figsize = (10,10))
plt.show()
```

```python
# Outliers Checking Section

# Describe the data to view statistical values
desc = dataset_new_after.describe()
desc
```

```python
# Determine new rows = deviation of 3 stdev from mean and +-1.5IQR for outliers checking
desc.loc['Stdev +3'] = desc.loc['mean'] + (desc.loc['std']*3)
desc.loc['Stdev -3'] = desc.loc['mean'] - (desc.loc['std']*3) 
desc.loc['+1.5IQR'] = (desc.loc['75%'] + (1.5 * abs(desc.loc['75%']) - desc.loc['25%']))
desc.loc['-1.5IQR'] = (desc.loc['25%'] - (1.5 * abs(desc.loc['75%']) - desc.loc['25%']))
desc
```

```python
# Check for outliers first method (Zscore +- 3 stdev)
from scipy import stats
dataset_new_after[(np.abs(stats.zscore(dataset_new_after)) > 3).any(axis=1)]
```

```python
# Create new dataframe to search for IQR outliers, second method.
IQR_pos = {i:desc.loc['+1.5IQR', i] for i in desc.columns}
IQR_neg = {i:desc.loc['-1.5IQR', i] for i in desc.columns}
IQR_df_positive = pd.DataFrame(IQR_pos, index=dataset_new_after.index)
IQR_df_negative = pd.DataFrame(IQR_neg, index=dataset_new_after.index)

IQR_df_positive.head()
```

```python
# Check outlier +1.5IQR and -1.5IQR data
dataset_new_after[(dataset_new_after > IQR_df_positive).any(axis=1)]
dataset_new_after[(dataset_new_after < IQR_df_negative).any(axis=1)]

# Result > 1.5IQR is year 1969 and 2001
# Result < 1.5IQR is year 1998

# Year 1998 and 2001 as obtained from both outliers finding method indicate the major event happened during those years. 1998 marked as asian financial crisis and 2001 Korea's economic difficulties due to unemployment and difficulties on business and industry.

# No outliers will be dropped in which both analyzed as essential that determined the Korea's GDP growth.
```

```python
# Building the regression model with sklearn.LinearRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Determine independent variables, as dataframe
x = dataset_new_after.drop('GDP growth', axis=1)
y = dataset_new_after[['GDP growth']]

# Split data to train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
Lr = LinearRegression()
Lr.fit(x_train, y_train)
```

```python
# See the coefficient (coef_) and constant/noise (intercept_)
print('Constant =', float(Lr.intercept_))

for i in range(len(x.columns)):
    print('Coefficient for', x.columns[i], 'is', Lr.coef_[0][i])
```

```python
# Lets check the prediction using the model and compare to the actual value
predict = Lr.predict(x_test)
compare = pd.DataFrame({'Prediction':predict.reshape(10), 'Actual':np.array(y_test).reshape(10)}, index=y_test.index)

compare
```

```python
''''
Alright, now lets evaluate the model using statsmodels library to check the rest of the assumptions;
1. The regression residuals must be normally distributed
2. The residuals must be homoskedastic, intercept value remain constant
3. No autocorrelation, no relation between regression residuals
'''

# For the variables
x1 = sm.add_constant(x)
model = sm.OLS(y,x1)
res = model.fit()

# Print temporary summary
res.summary()
```

```python
# Check the distribution plot of residuals (assumed to be normal distributed)
plt.figure(figsize=(5,5))
plt.tight_layout()
sbn.distplot(res.resid)
plt.show
```

```python
# Check residuals compared to theoritical quantile with pylab plotting
import pylab
sm.qqplot(res.resid, line='s')
pylab.show()
# As well print the mean of residuals, to see whether it is close to 0 or not
print('The mean of residuals =', np.mean(np.array(res.resid)))
```

```python
'''
Heteroskedasticity test, check whether the variance of error (intercept_) have a constant value (Homo), or not (Hetero).
White's heteroskedasticity test.
Null hypothesis Ho : variance of error have a constant value (Homo)
Alternative hpy Ha : variance of error have no constant value (Hetero)
The goal is to fail rejecting Ho or p-value > .05
'''
from statsmodels.stats import diagnostic as diag
lag_multi, pvalue, fvalue, f_pvalue = diag.het_white(res.resid, res.model.exog)

# Both p-value, lagrange multiplier test and f-statistics > .05, fail to reject Ho, Ha proposed which is homoskedastic.
pvalue, f_pvalue
```

```python
'''
Test for autocorrelation between residual, Ljung-Box test.
Ho : The data doesn't have any autocorrelation; random.
Ha : The data does indeed exist autocorrelation; not random / related
The goal is to fail rejecting Ho.
'''

# Set lags = min(10, number_of_obs//5)
lag_value = min(10, (len(x)))

# Ljung-Box Test
results = diag.acorr_ljungbox(res.resid, lags=lag_value)
ib_value, p_value = results
min(p_value)

# Minimum p-value is 0.1 which is > .05 (significance level), we fail to reject Ho, Ha proposed. There is no autocorrelation in residuals
```

```python
# The autocorrelation can be plotted with statsmodels.graphics library
sm.graphics.tsa.plot_acf(res.resid)
plt.show()

# The light blue area, confidence level 95%
```

```python
'''
Error calculations.
1. Mean Absolute Error(MAE), abs value of errors
2. Mean Squared Error (MSE), mean of the squared errors
3. Root Mean Squared Error (RMSE), root of MSE
Calculate using function from sklearn.metrics, actual = y_test, predict = predict
'''
from sklearn.metrics import mean_squared_error, mean_absolute_error
print('MAE =', mean_absolute_error(y_test, predict))
print('MSE =', mean_squared_error(y_test, predict))
print('RMSE =',np.sqrt(mean_squared_error(y_test, predict)))
```

```python
# Confidence interval, show range that if another 100 sample taken from population, approximately 95 samples would fall in the interval, termed as confidence level of 95%.
# Confidence interval _sm_OLS_fit.conf_int(alpha=.05 by default)
res.conf_int().rename(columns={0:'Min', 1:'Max'})
```

```python
'''
Review the statsmodels summary and hypothesis testing
Ho : There is no relation between explanatory and response variable; a.k.a independent and dependent var.
Ha : There is relation.
The goal is to reject Ho, to show that there is relation and the coeff of explanatory var. != 0
'''
# Check the summary again now
res.summary()

#P-values can be inferred from P>|t|
```

```python
# From the summary we can determine the insignificant (p-val>.05) regressor. Those variables will be removed and only leave variable with p-val<.05

# Remove : Gross cap formation, Broad money growth, and Unemployment. Lets try remove them from the equation.
x_res_new = x.drop(['Broad money growth','Gross cap formation', 'Unemployment'], axis=1)
y_res_new = y

# Review the summary after dropping them
x_res_new1 = sm.add_constant(x_res_new)
model_new = sm.OLS(y_res_new, x_res_new1)
res_new = model_new.fit()
res_new.summary()
```
