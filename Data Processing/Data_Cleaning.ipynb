```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('D:\Belajar\Project Py\Dataset\DataBank.csv')
dataset.head()
```

```python
# Transpose the dataframe
dataset = pd.DataFrame(dataset.transpose())
dataset
```

```python
# Assign the columns name equal to first row
dataset.columns = dataset.loc['Series Name', :]
dataset.head()
```

```python
# Drop 4 top rows, not part of data we need
dataset.drop(['Series Name', 'Series Code', 'Country Name', 'Country Code'], axis=0, inplace=True)
dataset
```

```python
# Sort by year ascending
dataset.sort_index(ascending=True, inplace=True)
dataset
```

```python
# Rename columns name, simplified
dataset.rename(columns={'Series Name':'year', 'GDP (current US$)':'GDP', 'Cereal production (metric tons)':'cereal production', 'Stocks traded, total value (% of GDP)':'total traded stock'}, inplace=True)
dataset.head()
```

```python
# Drop any NaN values which are useless in this case
dataset.dropna(axis=0, how='any', inplace=True)
dataset
```

```python
# Reassign the index label and index column.
dataset.index = [dataset.index[i].split(' ')[0] for i in range(len(dataset.index))]
dataset.columns.rename('year', inplace=True)

# Drop cereal production column, though I kinda curious..
dataset = dataset.drop('cereal production', axis=1).astype(float)
dataset
```

```python
# Our data is now cleaned, import more necessary libraries
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.linear_model import LinearRegression

# Our independent var. (x) wil be total traded stock and dependent var. (y) will be GDP
# Set y and x
y = dataset[['GDP']]
x = dataset[['total traded stock']]

# Set up regression model
Lr = LinearRegression()
Lr.fit(x,y)

# Plot
plt.scatter(x,y,c='cadetblue')
plt.plot(x, Lr.predict(x), label='Regression Line')
plt.title('Total traded stock vs GDP')
plt.ylabel('GDP in billion $')
plt.xlabel('Total traded stock % of GDP')
plt.legend()
plt.show()
```
